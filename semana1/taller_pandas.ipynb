# Taller básico pandas - numpy

```python
import pandas as pd
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos
cancion=pd.Series(['save your tears','no tears left to cry','kill this love'])
artista=pd.Series(['The weekend','Ariana Grande','BlackPink'])
duracion=pd.Series([300,200,100])
df_album=pd.DataFrame({'Nombre_cancion':cancion,'artista':artista,'Duracion':duracion})

df_album



```

```python
# Imprimir la cola del dataframe
df_album.tail()
```

```python
# Acceder a la columna duracion
df_album['Duracion']
```

```python
# Calcular el promedio de las duraciones
df_album.describe()
promedio=df_album['Duracion'].mean()
print("El promedio de duracion es:",promedio)
```

```python
# Obtener los nombres de las canciones que tengan duracion mayor que el promedio

canciones_mayores_que_promedio = df_album[df_album['Duracion'] > promedio]

nombres_canciones = canciones_mayores_que_promedio['Nombre_cancion']

print("Canciones con duración mayor que el promedio:")
print(nombres_canciones.tolist())

```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor

df_album['mayor_3_minutos']=pd.Series(df_album['Duracion'].apply(lambda T:T >promedio))
df_album
```

```python
# Convertir el dataframe a numpy array
import numpy as np
array=df_album.to_numpy()
array
```

```python
# Imprimir el shape del array
print(array.shape)
```

```python
# Hacer un reshape para convertir el arreglo de 3x3 a un arreglo de 12 elementos
array_convertido=array.reshape(12,)
array_convertido
```
